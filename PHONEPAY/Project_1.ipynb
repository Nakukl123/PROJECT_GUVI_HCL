```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Luxury_Housing_Bangalore.csv")
df.head()
```

```python
df.columns
```

```python
p = df['Ticket_Price_Cr']
p
```

```python
df['Ticket_Price_Cr'] = (
    pd.to_numeric(
        df['Ticket_Price_Cr']
        .astype(str)
        .str.replace('₹', '', regex=False)
        .str.replace('Cr', '', regex=False)
        .str.replace('C', '', regex=False)
        .str.strip(),
        errors='coerce'
    )
    .round(2)
)
```

```python
df.head()
```

```python
df['Amenity_Score'] = df['Amenity_Score'].fillna(df['Amenity_Score'].median())
```

```python
df['Possession_Status'].isna().any()
```

```python
text_cols = ['Developer_Name', 'Micro_Market']

for col in text_cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()                      # remove leading/trailing spaces
        .str.replace(r'\s+', ' ', regex=True)  # remove extra spaces
        .str.title()                      # normalize case
    )
```

```python
df.head()
```

```python
df.loc[df['Unit_Size_Sqft'] < 0, 'Unit_Size_Sqft'] = pd.NA
df['Unit_Size_Sqft'] = df['Unit_Size_Sqft'].fillna(df['Unit_Size_Sqft'].median())
(df['Unit_Size_Sqft'] <= 0).sum()
```

```python
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].abs()
```

```python
df['Price_per_Sqft'] = (
    (df['Ticket_Price_Cr'] * 1_0000_0000) / df['Unit_Size_Sqft']
).round(0)
```

```python
Quarter = df['Purchase_Quarter']
Quarter
```

```python
df['Purchase_Quarter'] = pd.to_datetime(df['Purchase_Quarter'], errors='coerce').dt.date
```

```python
df['Quarter_Number'] = pd.to_datetime(df['Purchase_Quarter'], errors='coerce').dt.quarter
```

```python
df['Booking_Flag'] = df['Transaction_Type'].notna().astype(int)
```

```python
df.isna().any().any()
```

```python
df.isna().sum()
```

```python
df['Unit_Size_Sqft'] = df['Unit_Size_Sqft'].fillna(df['Unit_Size_Sqft'].median())
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].fillna(df['Ticket_Price_Cr'].median())
df['Price_per_Sqft'] = df['Price_per_Sqft'].fillna(df['Price_per_Sqft'].median())
df['Buyer_Comments'] = df['Buyer_Comments'].fillna('No Comment')
```

```python
df[df['Unit_Size_Sqft'] < 0]
```

```python
df[df['Price_per_Sqft'] < 0]
```

```python
df['Configuration'] = (
    df['Configuration']
    .astype(str)
    .str.strip()
    .str.upper()
    .str.replace(' ', '', regex=False)
)
```

```python
df.to_csv('clean_real_estate_data.csv', index=False)
```
